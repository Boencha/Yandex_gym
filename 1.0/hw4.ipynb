```python
#A. Словарь синонимов
d = {}
for i in range(int(input())):
    w1, w2 = input().split()
    d[w1] = w2
    
x = input()

for w1, w2 in d.items():
    if w1 == x:
        print(w2)
    elif w2 == x:
        print(w1)
```

```python
#B. Номер появления слова
with open('input.txt') as inp:
    d, lines = {}, inp.readlines()
    for line in lines:
        for word in line.split():
            print(d.setdefault(word, 0), end = ' ')
            d[word] = d.get(word, 0) + 1
```

```python
#C. Самое частое слово
d = {}
with open('input.txt') as inp:
    lines = inp.readlines()
    for line in lines:
        for w in line.split():
            d[w] = d.get(w, 0) + 1
            
m_v, ans = max(d.values()), []
for k, v in d.items():
    if v == m_v:
        ans.append(k)
        
print(sorted(ans)[0])
```

```python
#D. Клавиатура
n = int(input())
arr = list(map(int, input().split()))
keyboard = {k + 1: arr[k] for k in range(n)}

s = int(input())
arr_1 = list(map(int, input().split()))
for i in range(s):
    keyboard[arr_1[i]] = keyboard.get(arr_1[i]) - 1

for v in keyboard.values():
    if v < 0:
        print('YES')
    else:
        print('NO')
```

```python
#E. Пирамида
cubes = {}
for i in range(int(input())):
    w, h = map(int, input().split())
    if w not in cubes:
        cubes[w] = h
    elif w in cubes and cubes[w] < h:
        cubes[w] = h
 
ans = 0       
for h in cubes.values():
    ans += h
    
print(ans)
```

```python
#F. Продажи
clients = {}
with open('input.txt') as inp:
    lines = inp.readlines()
    for line in lines:
#for i in range(int(input())):
        #line = input()
        client, product, сount = line.split()
        if client not in clients:
            clients[client] = {product: int(сount)}
        else:
            if product not in clients[client]:
                clients[client][product] = int(сount)
            else:
                clients[client][product] += int(сount)
    
for client in sorted(clients.keys()):
    print(f'{client}:')   
    for product in sorted(clients[client].keys()):
        print(f'{product} {clients[client][product]}')
                
```

```python
#G. Банковские счета
bank = {}
ans = []
with open('input.txt') as inp:
    lines = inp.readlines()
    for line in lines:
#for i in range(int(input())):
        #line = input()
        operation, *infa = line.split()
        if operation == 'DEPOSIT':
            client, sum = infa[0], infa[1]
            bank[client] = bank.setdefault(client, 0) + int(sum)
            
        elif operation == 'WITHDRAW':
            client, sum = infa[0], infa[1]
            bank[client] = bank.setdefault(client, 0) - int(sum)
            
        elif operation == 'BALANCE':
            client = infa[0]
            ans.append(bank.get(client, 'ERROR'))
            
        elif operation == 'TRANSFER':
            client1, client2, sum = infa[0], infa[1], int(infa[2])
            bank[client2] = bank.setdefault(client2, 0) + sum
            bank[client1] = bank.setdefault(client1, 0) - sum
            
        elif operation == 'INCOME':
            procent = int(infa[0]) / 100
            for client, sum in bank.items():
                if sum > 0:
                    bank[client] = bank[client] + int(bank[client] * procent)
                    
for i in ans:
    print(i)
```

```python
#I. Контрольная по ударениям
def not_in_d(word):
    flag, mj = True, 0
    for l in word:
        if l.isupper():
            mj += 1
    if mj != 1:
        flag = False
    return flag


ans, d= 0, {}
for i in range(int(input())):
    w = input()
    if w.lower() not in d:
        d[w.lower()] = [w]
    else:
         d[w.lower()].append(w)        


for word in input().split():
    if word.lower() not in d:
        if not_in_d(word) == False:
            ans += 1    
    else:
        fl = False
        for i in d[word.lower()]:
                if i == word:
                    fl = True
        if fl == False:
                ans += 1

                
print(ans)
```
