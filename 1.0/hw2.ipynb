```python
#A. Возрастает ли список?
arr = list(map(int, input().split()))

flag = 1
for i in range(len(arr) - 1):
    if arr[i] >= arr[i + 1]:
        flag = 0
        
if flag:
    print('YES')
else:
    print('NO')
```

```python
#B. Определить вид последовательности
flag = 1
arr = []
while flag:
    x = int(input())
    if x == -2000000000:
        flag = False
    else:
        arr.append(x)

point = 0       
for i in range(len(arr) - 1):
    if arr[i] == arr[i + 1]:
        if point == 0 or point == 1:
            point = 1
        elif point == 2 or point == 3:
            point = 3
        elif point == 4 or point == 5:
            point = 5
        else:
            point = 6
            
    elif arr[i] < arr[i + 1]:
        if point == 0 or point == 2:
            point = 2
        elif point == 1 or point == 3:
            point = 3
        else:
            point = 6
            
    elif arr[i] > arr[i + 1]:
        if point == 0 or point == 4:
            point = 4
        elif point == 1 or point == 5:
            point = 5
        else:
            point = 6
        
if point == 1:
    print('CONSTANT')
    
elif point == 2:
    print('ASCENDING')
    
elif point == 3:
    print('WEAKLY ASCENDING')
    
elif point == 4:
    print('DESCENDING')
    
elif point == 5:
    print('WEAKLY DESCENDING')
    
else:
    print('RANDOM')
```

```python
#D. Больше своих соседей
arr = list(map(int, input().split()))

count = 0
for i in range(1,len(arr) - 1):
    if arr[i - 1] < arr[i] and arr[i] > arr[i + 1]:
        count += 1
        
print(count)
```

```python
#G. Наибольшее произведение двух чисел
array = list(map(int, input().split()))

if len(array) == 2:
    print(min(array), max(array))
else:
    max_1 = max(array)
    array.pop(array.index(max_1))
    max_2 = max(array)

    min_1 = min(array)
    array.pop(array.index(min_1))
    min_2 = min(array)
    
    if max_1 * max_2 > min_1 * min_2:
        print(max_2, max_1)
    else:
        print(min_1, min_2)
```

```python
#H. Наибольшее произведение трех чисел
array = list(map(int, input().split()))

if len(array) == 3:
    print(*array)
else:
    max_1 = max(array)
    array.pop(array.index(max_1))
    max_2 = max(array)
    array.pop(array.index(max_2))
    max_3 = max(array)

    min_1 = min(array)
    array.pop(array.index(min_1))
    min_2 = min(array)
    
    if max_1 * max_2 * max_3 > min_1 * min_2 * max_1:
        print(max_2, max_1, max_3)
    else:
        print(min_1, min_2, max_1)
```

```python
#I. Сапер
n, m, am = list(map(int, input().split()))
mines = [[int(i) for i in input().split()] for _ in range(am)]

setka = [[0] * (m + 2) for _ in range(n + 2)]

det_i = [0, 0, 1, 1, 1, -1, -1, -1]
det_j = [1, -1, -1, 0, 1, -1, 0, 1]


for min in range(am):
    setka[mines[min][0]][mines[min][1]] = '*'
    for k in range(8):
        if setka[mines[min][0]+ det_i[k]][mines[min][1] + det_j[k]] != '*':
            setka[mines[min][0]+ det_i[k]][mines[min][1] + det_j[k]] += 1
    
            
for i in range(1, n + 1):
    for j in range(1, m + 1):
        print(setka[i][j], end=' ')
    print()
```

```python
#J. Треугольник Максима
n = int(input())
notion = [input().split() for _ in range(n)]
dist = [float(30), float(4000)]


for i in range(1, n):
    if float(notion[i][0]) > float(notion[i -1][0]):
        if notion[i][1] == 'further':
            dist[1] = min(dist[1], (float(notion[i][0]) + float(notion[i - 1][0])) / 2)
        else:
            dist[0] = max(dist[0], (float(notion[i][0]) + float(notion[i - 1][0])) / 2)
        
    if float(notion[i][0]) < float(notion[i -1][0]):
        if notion[i][1] == 'further':
            dist[0] = max(dist[0], (float(notion[i][0]) + float(notion[i - 1][0])) / 2)
        else:
            dist[1] = min(dist[1], (float(notion[i][0]) + float(notion[i - 1][0])) / 2)
             
print(*dist)
```
