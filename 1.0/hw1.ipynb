```python
#A. Кондиционер
room, cond = list(map(int, input().split()))
mode = input()
ans = ''
if mode == 'freeze':
    if room > cond:
        ans = cond
    else:
        ans = room
        
elif mode == 'heat':
    if room < cond:
        ans = cond
    else:
        ans = room
    
elif mode == 'auto':
    ans = cond
    
elif mode == 'fan':
    ans = room
    
print(ans)
```

```python
#B. Треугольник
st = [int(input()) for _ in range(3)]
flag = 1

for i in range(3):
    if st[i] >= sum(st) - st[i]:
       flag = 0

if flag == 1:      
    print('YES')
else:
  print('NO')
```

```python
#C. Телефонные номера
add = list(input())

contact = [list(input())  for _ in range(3)]
add = [char for char in add if char not in ('+', '(', ')', '-')]

if len(add) == 7:
    add = ['7', '4', '9', '5'] + add
   
for i in range(3):
    contact[i] = [sim for sim in contact[i] if sim not in ('+', '(', ')', '-')]
    if len(contact[i]) == len(add) - 4:
        contact[i] = ['4', '9', '5'] + contact[i]
        if contact[i] == add[1 :]:
            print('YES')
        else:
            print('NO')
            
            
    elif len(contact[i]) == len(add):
        if contact[i][1:] == add[1:]:
            print('YES')
        else:
            print('NO')
            
    else:
        print('NO')
```

```python
#D. Уравнение с корнем
a, b, c = int(input()), int(input()), int(input())

if c < 0:
    print('NO SOLUTION')
    
else:
    if a != 0: 
        x = (c ** 2 - b) / a
        flag = x.is_integer()
        if flag :
            print(int(x))
        else:
            print('NO SOLUTION')
    else:
        if b == c == 0 or b == c ** 2:
            print('MANY SOLUTIONS')
            
        else:
            print('NO SOLUTION')
```

```python
#F. Расстановка ноутбуков
a_1, b_1, a_2, b_2 = list(map(int, input().split()))

x_1 = max(a_1, b_1)
y_1 = min(a_1, b_1)
x_2 = max(a_2, b_2)
y_2 = min(a_2, b_2)

if  x_2 <= y_1:
    print(x_1 + y_2, y_1)
    
elif x_1 <= y_2:
    print(x_2 + y_1, y_2)
    
else:
    print(max(x_1, x_2), y_1 + y_2)
```

```python
#G. Детали
n, k, m =list(map(int, input().split()))

zag = int(n / k)
det = zag * int(k / m)

ost = (n % k) + zag * (k % m)

if k >= m:
    while ost >= k:
        zag = 0
        zag += int(ost / k) #кол-во заготовок
        ost -= zag * k #оставшийся остаток после изг-я заготовок
        det += zag * int(k / m)
        ost += zag * (k % m)
    print(det)
    
else:
    print(0)
```

```python
#H. Метро
a, b, n, m = int(input()), int(input()), int(input()), int(input())

max_1 = a * (n + 1) + n
max_2 = b * (m + 1) + m
max_0 = min(max_1, max_2)

min_1 = n + a * (n - 1)
min_2 = m + b * (m - 1)
min_0 = max(min_1, min_2)

if max_0 < min_0:
    print('-1')
    
else:
    print(min_0, max_0)
```

```python
#I. Узник замка Иф
size = [int(input()) for _ in range(5)]

min_st = size[0]
min_st_2 = size[1]
for i in range(1, 3):
    if size[i] < min_st:
        min_st_2 = min_st
        min_st = size[i]
    if size[i] < min_st_2:
        min_st_2 = size[i]
        
if min(min_st, min_st_2) <= min(size[3], size[4]) and max(min_st, min_st_2) <= max(size[3], size[4]):
    print('YES')
else:
    print('NO')
```
