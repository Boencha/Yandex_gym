```python
#G
from itertools import combinations

num_points = int(input())
coordinates = [tuple(map(int, input().split())) for _ in range(num_points)]

def form_square(pt1, pt2, pt3, pt4):
    pts = [pt1, pt2, pt3, pt4]
    dist_squared = sorted([sum((x - y) ** 2 for x, y in combinations(pts, 2))])
    return dist_squared[0] == dist_squared[1] == dist_squared[2] == dist_squared[3] and dist_squared[4] == dist_squared[5]


min_additional_points = float('inf')
for comb in combinations(coordinates, 4):
    if form_square(*comb):
        min_additional_points = min(min_additional_points, 0)
    else:
        min_additional_points = min(min_additional_points, 4 - sum(pt in comb for pt in coordinates))


print(int(min_additional_points))
if min_additional_points > 0:
    for _ in range(int(min_additional_points)):
        print("0 0")
```

```python
#H
def compute_vector(coords: list[int]):
    x1, y1, x2, y2 = coords
    start_x, start_y = x1, y1
    end_x, end_y = x2, y2
    if start_y > end_y or (start_y == end_y and start_x > end_x):
        start_x, start_y = x2, y2
        end_x, end_y = x1, y1
    delta_x, delta_y = end_x - start_x, end_y - start_y
    return (start_x, start_y), (delta_x, delta_y)

def identify_vector_directions(shapes: list[list[int]]):
    direction_map = dict()
    for coords in shapes:
        (start_x, start_y), (delta_x, delta_y) = compute_vector(coords)
        if (delta_x, delta_y) not in direction_map:
            direction_map[(delta_x, delta_y)] = []
        direction_map[(delta_x, delta_y)].append((start_x, start_y))
    return direction_map

def find_min_points_to_align(num_shapes: int, shape_A: list[list[int]], shape_B: list[list[int]]) -> int:
    directions_A = identify_vector_directions(shape_A)
    directions_B = identify_vector_directions(shape_B)

    difference_counts = dict()
    max_overlap = 0
    for key, items_B in directions_B.items():
        if key not in directions_A:
            continue
        for (x1, y1) in items_B:
            for (x2, y2) in directions_A[key]:
                delta_x, delta_y = x2 - x1, y2 - y1
                if (delta_x, delta_y) not in difference_counts:
                    difference_counts[(delta_x, delta_y)] = 0
                difference_counts[(delta_x, delta_y)] += 1
                if max_overlap < difference_counts[(delta_x, delta_y)]:
                    max_overlap = difference_counts[(delta_x, delta_y)]
    return num_shapes - max_overlap

def run_tests():
    test1 = find_min_points_to_align(
        num_shapes=5,
        shape_A=[[0, 0, 1, 2], [1, 0, 0, 2], [2, 0, 2, 2], [4, 0, 3, 2], [4, 0, 5, 2]],
        shape_B=[[9, -1, 10, 1], [10, 1, 9, 3], [8, 1, 10, 1], [8, 1, 9, -1], [8, 1, 9, 3]],
    )
    assert test1 == 3, test1

    test2 = find_min_points_to_align(
        num_shapes=1,
        shape_A=[[3, 4, 7, 9]],
        shape_B=[[-1, 3, 3, 8]],
    )
    assert test2 == 0, test2

    test3 = find_min_points_to_align(
        num_shapes=1,
        shape_A=[[-4, 5, 2, -3]],
        shape_B=[[-12, 4, -2, 4]],
    )
    assert test3 == 1, test3

if __name__ == '__main__':
    debug = True
    if debug:
        run_tests()
    num_points = int(input())
    shape_A = []
    for _ in range(num_points):
        shape_A.append(list(map(int, input().split())))
    shape_B = []
    for _ in range(num_points):
        shape_B.append(list(map(int, input().split())))
    result = find_min_points_to_align(num_points, shape_A, shape_B)
    print(result)
```

```python
#J
import math

class Transmission:
    def __init__(self, source, target, data_id):
        self.source = source
        self.target = target
        self.data_id = data_id

    def __str__(self):
        return 'Transmission: source {} target {} (data_id: {})'.format(self.source.device_id, self.target.device_id, self.data_id)

class Device:
    def __init__(self, total_devices, device_id):
        self.received_data = set()
        self.device_id = device_id
        self.priority = {
            other_id: 0 for other_id in range(total_devices) if other_id != device_id
        }

    def __str__(self):
        return 'Device {}. received_data {}. priority {} '.format(self.device_id, self.received_data, self.priority)

def find_rare_data_packet(device: Device, data_packets: list[set]):
    rarest_packet = None
    smallest_size = math.inf
    for data_id, packet in enumerate(data_packets):
        if data_id in device.received_data:
            continue
        if len(packet) < smallest_size:
            smallest_size = len(packet)
            rarest_packet = data_id
        elif len(packet) == smallest_size and rarest_packet > data_id:
            rarest_packet = data_id
    return rarest_packet

def select_sending_device(devices: list[Device], data_packets: list, rarest_packet_id: int):
    min_received = math.inf
    selected_device = None
    for device_id in data_packets[rarest_packet_id]:
        current_device = devices[device_id]
        if len(current_device.received_data) < min_received:
            min_received = len(current_device.received_data)
            selected_device = current_device
        elif len(current_device.received_data) == min_received and len(selected_device.received_data) > len(current_device.received_data):
            min_received = len(current_device.received_data)
            selected_device = current_device
        elif len(current_device.received_data) == min_received and len(selected_device.received_data) == len(current_device.received_data) and selected_device.device_id > device_id:
            min_received = len(current_device.received_data)
            selected_device = current_device
    return selected_device

def choose_best_request(devices: list[Device], pending_requests: list[Transmission], target_device_id: int):
    highest_priority = -math.inf
    best_request = None
    for req in pending_requests:
        priority_info = devices[req.target.device_id].priority
        if priority_info[req.source.device_id] > highest_priority:
            best_request = req
            highest_priority = priority_info[req.source.device_id]
        elif priority_info[req.source.device_id] == highest_priority and len(req.source.received_data) < len(best_request.source.received_data):
            best_request = req
            highest_priority = priority_info[req.source.device_id]
        elif priority_info[req.source.device_id] == highest_priority and len(req.source.received_data) == len(best_request.source.received_data) and best_request.source.device_id > req.source.device_id:
            best_request = req
            highest_priority = priority_info[req.source.device_id]
    return best_request

def devices_process_requests(devices: list[Device], data_packets: list, requests: dict, response: list[int]):
    best_requests = []
    for device_id, pending_requests in requests.items():
        if not pending_requests:
            continue
        best_request: Transmission = choose_best_request(
            devices=devices,
            pending_requests=pending_requests,
            target_device_id=device_id
        )
        best_requests.append(best_request)

    for best_request in best_requests:
        devices[best_request.source.device_id].priority[best_request.target.device_id] += 1
        devices[best_request.source.device_id].received_data.add(best_request.data_id)
        data_packets[best_request.data_id].add(best_request.source.device_id)

def devices_initiate_requests(devices: list[Device], data_packets: list, requests: dict, response: list[int]):
    for device in devices:
        if device.device_id == 0:
            continue
        rarest_packet_id: int = find_rare_data_packet(
            device=device,
            data_packets=data_packets,
        )
        if rarest_packet_id is None:
            continue
        sending_device = select_sending_device(
            devices=devices,
            data_packets=data_packets,
            rarest_packet_id=rarest_packet_id,
        )
        new_request = Transmission(
            source=device,
            target=sending_device,
            data_id=rarest_packet_id,
        )
        requests[sending_device.device_id].append(new_request)

def update_response(devices: list[Device], data_packets: list, response: list[int]):
    for device in devices:
        if len(device.received_data) == len(data_packets):
            continue
        response[device.device_id] += 1

def advance_time_slot(devices: list[Device], data_packets: list, requests: dict, response: list[int]):
    update_response(devices, data_packets, response)
    devices_initiate_requests(devices, data_packets, requests, response)
    devices_process_requests(devices, data_packets, requests, response)

def solve_problem(num_devices, num_data_packets) -> list[int]:
    devices = [
        Device(
            total_devices=num_devices,
            device_id=i,
        )
        for i in range(num_devices)
    ]

    devices[0].received_data = set(range(num_data_packets))
    data_packets = [
        {0} for _ in range(num_data_packets)
    ]

    response = [0 for _ in range(num_devices)]
    while any(len(packet) != num_devices for packet in data_packets):
        requests = {
            i: [] for i in range(num_devices)
        }
        advance_time_slot(devices, data_packets, requests, response)
    print(*response[1:])

def run_tests():
    pass

if __name__ == '__main__':
    is_debug = True
    if is_debug:
        run_tests()
    num_devices, num_data_packets = list(map(int, input().split()))
    solve_problem(num_devices, num_data_packets)
```
