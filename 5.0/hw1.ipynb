```python
#D
import copy

def __reformat_table_str_to_list(text_table: str) -> list[list[str]]:
    new_format_table: list[list[str]] = []
    for line in text_table.split():
        new_format_table.append(list(x for x in line))
    return new_format_table


board_: list[str] = []
for _ in range(8):
    board_.append(input())
board = __reformat_table_str_to_list(text_table='\n'.join(board_))


board_copy: list[list[str]] = copy.deepcopy(board)
rock = 'R'
bishop = 'B'
attack = 'X'
positions_all_figures: set[tuple[int, int]] = set()
result = 0


for y in range(8):
    for x in range(8):
        if board_copy[y][x] == rock:
            positions_all_figures.add((x, y))
            for x1 in range(x + 1, 8, 1):
                if board_copy[y][x1] in {rock, bishop}:
                    break
                elif board_copy[y][x1] == attack:
                    continue
                else:
                    board_copy[y][x1] = attack
                    result += 1
            for x1 in range(x - 1, -1, -1):
                if board_copy[y][x1] in {rock, bishop}:
                    break
                elif board_copy[y][x1] == attack:
                    continue
                else:
                    board_copy[y][x1] = attack
                    result += 1
            for y1 in range(y + 1, 8, 1):
                if board_copy[y1][x] in {rock, bishop}:
                    break
                elif board_copy[y1][x] == attack:
                    continue
                else:
                    board_copy[y1][x] = attack
                    result += 1
            for y1 in range(y - 1, -1, -1):
                if board_copy[y1][x] in {rock, bishop}:
                    break
                elif board_copy[y1][x] == attack:
                    continue
                else:
                    board_copy[y1][x] = attack
                    result += 1

for y in range(8):
    for x in range(8):
        if board_copy[y][x] == bishop:
            positions_all_figures.add((x, y))
            for i in range(1, 8, 1):
                x1 = x + i
                y1 = y + i
                if x1 < 0 or x1 >= 8 or y1 >= 8 or y1 < 0:
                    continue
                if (x == x1 and y == y1) or board_copy[y1][x1] == attack:
                    continue
                if board_copy[y1][x1] in {rock, bishop}:
                    break
                else:
                    board_copy[y1][x1] = attack
                    result += 1
            for i in range(0, -8, -1):
                x1 = x + i
                y1 = y + i
                if x1 < 0 or x1 >= 8 or y1 >= 8 or y1 < 0:
                    continue
                if (x == x1 and y == y1) or board_copy[y1][x1] == attack:
                    continue
                if board_copy[y1][x1] in {rock, bishop}:
                    break
                else:
                    board_copy[y1][x1] = attack
                    result += 1
            for i in range(0, -8, -1):
                x1 = x + i
                y1 = y - i
                if x1 < 0 or x1 >= 8 or y1 >= 8 or y1 < 0:
                    continue
                if (x == x1 and y == y1) or board_copy[y1][x1] == attack:
                    continue
                if board_copy[y1][x1] in {rock, bishop}:
                    break
                else:
                    board_copy[y1][x1] = attack
                    result += 1
            for i in range(0, 8, 1):
                x1 = x + i
                y1 = y - i
                if x1 < 0 or x1 >= 8 or y1 >= 8 or y1 < 0:
                    continue
                if (x == x1 and y == y1) or board_copy[y1][x1] == attack:
                    continue
                if board_copy[y1][x1] in {rock, bishop}:
                    break
                else:
                    board_copy[y1][x1] = attack
                    result += 1


for x, y in positions_all_figures:
    board_copy[y][x] = attack
    result += 1


print(8 * 8 - result)
```

```python
#E
import sys
sys.set_int_max_str_digits(1000000000)

def ans(n, k, d):
    prefix = n
    if k == 0:
        return -1
    for j in range(10):
        prefix_digit = prefix * 10 + j
        if prefix_digit % k == 0:
            if d - 1 > 0:
                return int(f'{prefix_digit}' + '0'*(d -1))
            else:
                return prefix_digit
    
    return -1


n, k, d = map(int, input().split())
answer = ans(n, k, d)
print(answer)
```

```python
#H
L, x1, v1, x2, v2 = list(map(int, input().split()))


def time_min_sovp(L, x1, v1, x2, v2):
    if v1 - v2 == 0:
        return -1
    
    def funct(n):
        return ((x1 - x2) + n * L) / (v2 - v1)
    
    results = []
    for x in range(-5000, 5000):
        value = funct(x)
        if value != -1 and value >= 0:
            results.append(value)
    
    if results:
        return min(results)
    return -1


def time_min_nesovp(L, x1, v1, x2, v2):
    if v1 + v2 == 0:
        return -1
    
    def funct(n):
        return (-(x1 + x2) + n * L) / (v2 + v1)
    
    results = []
    for x in range(-5000, 5000):
        value = funct(x)
        if value != -1 and value >= 0:
            results.append(value)
    
    if results:
        return min(results)
    return -1

if x1 == x2:
    print('YES')
    print(0)
else:
    res_1 = time_min_sovp(L, x1, v1, x2, v2)
    res_2 = time_min_nesovp(L, x1, v1, x2, v2)
    if res_1 == -1 and res_2 == -1:
        print('NO')
    elif res_1 != -1 and res_2 != -1:
        print('YES')
        print(min(res_1,res_2))
    elif res_1 == -1:
        print('YES')
        print(res_2)
    elif res_2 == -1:
        print('YES')
        print(res_1)
```

```python
#I
def is_leap_year(year):
    return year % 400 == 0 or (year % 4 == 0 and year % 100 != 0)

def get_month_number(month_name):
    months = {
        'January': 1,
        'February': 2,
        'March': 3,
        'April': 4,
        'May': 5,
        'June': 6,
        'July': 7,
        'August': 8,
        'September': 9,
        'October': 10,
        'November': 11,
        'December': 12
    }
    return months.get(month_name)

N = int(input())
year = int(input())
holidays = []
for _ in range(N):
    day, month_name = input().split()
    month = get_month_number(month_name)
    holidays.append((int(day), month))

first_day_of_year = input().strip()

def get_days_in_month(month, year):
    if month in {1, 3, 5, 7, 8, 10, 12}:
        return 31
    elif month in {4, 6, 9, 11}:
        return 30
    elif month == 2:
        return 29 if is_leap_year(year) else 28
    
#

days_count = {'Monday': 0, 'Tuesday': 0, 'Wednesday': 0, 'Thursday': 0, 'Friday': 0, 'Saturday': 0, 'Sunday': 0}

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
first_day_month = days.index(first_day_of_year)

for month in range(1, 13):
    days_in_month = get_days_in_month(month, year)
    for day in range(1, days_in_month + 1):
        day_of_week = (day + first_day_month - 1) % 7
        days_count[days[day_of_week]] += 1
    
    for holiday in holidays:
        day_h, month_h = holiday
        if month_h == month:
            day_of_week_h = (day_h + first_day_month - 1) % 7
            days_count[days[day_of_week_h]] -= 1

    first_day_month = day_of_week + 1
    

best_day = max(days_count, key=days_count.get)
worst_day = min(days_count, key=days_count.get)

print(best_day, worst_day)
```
